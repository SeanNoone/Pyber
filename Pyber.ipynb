```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as py
```

```python
ride_string = 'raw_data/ride_data.csv'
city_string = 'raw_data/city_data.csv'

ride_data_df = pd.read_csv(ride_string, encoding="iso-8859-1", low_memory=False)
city_data_df = pd.read_csv(city_string, encoding="iso-8859-1", low_memory=False)
```

```python
combined_uber_df = pd.merge(city_data_df, ride_data_df, on = 'city', how="outer").groupby(["city"],
                                                         as_index=True)
combined_piechart_df = pd.merge(city_data_df, ride_data_df, on = 'city', how="outer").groupby(["type"],
                                                         as_index=True)


combined_piechart2_df = pd.merge(city_data_df, ride_data_df, on = 'city', how="outer")
combined_piechart2_df = combined_piechart2_df.groupby(["type", "city"]).sum()
combined_piechart3_df = combined_piechart2_df.groupby(["type"]).sum()
```

```python
average_fare = combined_uber_df["fare"].mean()
print (average_fare)
```

```python
total_number_rides = combined_uber_df["city"].count()
print(total_number_rides)
```

```python
city_type = combined_uber_df["type"].unique()
print(city_type)
len(city_type)
```

```python
total_drivers = combined_uber_df["driver_count"].mean()
print(total_drivers)
```

```python
uber_df = pd.DataFrame({"Rides Per City":total_number_rides, "Average Fare":average_fare, "City Type":city_type, "Total Drivers":total_drivers})
uber_df
```

```python
rural_data = uber_df[uber_df["City Type"] == 'Rural']
urban_data = uber_df[uber_df["City Type"] == 'Urban']
suburban_data = uber_df[uber_df["City Type"] == 'Suburban']
```

```python
urban_data.head()
```

```python
rural_data.head()
```

```python
plt.grid()

urban_data_plt = plt.scatter(urban_data['Rides Per City'], urban_data['Average Fare'], marker="o", facecolors="gold", edgecolors="black", s=urban_data['Total Drivers']*10, 
             label= 'Urban', alpha=0.75)

suburban_data_plt = plt.scatter(suburban_data['Rides Per City'], suburban_data['Average Fare'], marker="o", facecolors="red", edgecolors="black", s=suburban_data['Total Drivers']*10, 
             label= 'Urban', alpha=0.75)

rural_data_plt = plt.scatter(rural_data['Rides Per City'], rural_data['Average Fare'], marker="o", facecolors="blue", edgecolors="black", s=rural_data['Total Drivers']*10, 
             label= 'Urban', alpha=0.75) 

plt.ylim(15,55)

plt.xlim(0,40)

plt.legend((urban_data_plt, suburban_data_plt, rural_data_plt), ("Urban, Suburban, Rural"), title="City Types")

plt.title("Uber Rides Sharing Data (2016)")
plt.xlabel("Average Fare $")
plt.ylabel("Total Number of Rides Per City")



plt.show()
```

```python
fares_label = ["Rural", "Suburban", "Urban"]
colors = ["red","lightskyblue","gold"]
explode = (0,0,0.1)
```

```python
total_fares = combined_piechart_df["fare"].sum()

plt.pie(total_fares, explode=explode, labels=fares_label, colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=140)

plt.title("% of Total Fares Per City Type")

plt.axis("equal")

plt.savefig("%rides_type")
plt.show()
```

```python
total_count = combined_piechart_df["fare"].count()

plt.pie(total_count, explode=explode, labels=fares_label, colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=140)

plt.title("% of Total Rides Per City Type")

plt.axis("equal")

plt.savefig("%rides_type")
plt.show()
```

```python
total_drivers = combined_piechart3_df["driver_count"]

plt.pie(total_drivers, explode=explode, labels=fares_label, colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=140)

plt.title("% of Total Drivers Per City Type")

plt.axis("equal")

plt.savefig("%rides_type")
plt.show()
```
